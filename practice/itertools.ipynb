# Itertools

```python
from itertools import groupby

family = [dict(name="Ajay", age=36), dict(name="Priti", age=32), dict(name="Saanvi", age=7), dict(name="Anant", age=3)]


print(f"family => \n {family}\n")

# Group the list of dictionaries
group_obj = groupby(family, key=lambda p: 'Adult' if p['age'] >= 18 else 'Kid')

print(f"group_obj => \n {group_obj}\n")

for key, value in group_obj:
    print(key, list(value))
```
