```python
# get dictionary with each number as key, and its square as value

# dict with indexed keys
my_num_squares = {x:x**2 for x in range(1,11)}
# dict with string keys
my_num_squares_string = {f'{x}':x**2 for x in range(1,11)}

print(f"my_num_squares => \n{my_num_squares}\n")
print(f"square of 5 (my_num_squares[5]) => \n{my_num_squares[5]}\n")
print(f"my_num_squares_string => \n{my_num_squares_string}\n")
print(f"square of 5 (my_num_squares_string['5']) => \n{my_num_squares_string['5']}\n")
```

```python
# get list of dict
[dict(num=x, sqr=x*x) for x in range(1,11)]
```
